# CVAR - Conditional Value at Risk

## Definition
Average loss when VaR is exceeded, mean average loss for worst scenarios. 

## Before CVaR we need the VaR
Here let's reuse what we did before.

```python
from finance_compute_helper import show_ptf_var_histo_report, ptf_yields_returns, compute_var_histo
from datetime import date

today = date.today().strftime("%Y-%m-%d")

portfolio_dict = {
    "CW8.PA": 50/100,
    "NVDA": 25/100,
    "TSLA": 25/100,
}

yield_returns, list_names = ptf_yields_returns(portfolio_dict,"2022-01-01",today)
var95 = compute_var_histo(yield_returns, 95)
show_ptf_var_histo_report(list_names, 95, yield_returns)
```

## CVaR
To computer the CVaR we need to get the financial yields returns inferiors to the VaR, and get the average.

```python
from finance_compute_helper import compute_cvar
print(str((compute_cvar(yield_returns, var95) * 100).round(2)) + '%')
```
